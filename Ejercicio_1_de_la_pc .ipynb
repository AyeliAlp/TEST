```python
import pandas as pd

LibrosDFs=pd.read_html(LibrosLink, # link
                        header=0, # where is the header?
                        flavor='bs4')
```

```python
#Quería saber el tipo de datos que tenia y use la funcion type 
type(LibrosDFs)
```

```python
#Comprobé el tamaño
len(LibrosDFs)
```

```python
#Hice una copia de los datos y seleccioné la tabla que voy a limpiar de todas las tablas que habian en la pagina web "[4]"
Libros=LibrosDFs[4].copy()
Libros
```

```python
Libros.info() #Queria saber el tipo de variable de cada una de las columnas, asi que lo averigué con la funcion ".info()"
```

```python
Libros.rename(columns={"Ventas aproximadas": "Ventas_Aproximadas"}) #Renombré a la columna de Ventas aproximadas para que 
#posteriormente los espacios en blanco en el nombre no sean un problema.
```

```python
LibrosArreglo=Libros.rename(columns={"Ventas aproximadas": "Ventas_Aproximadas"}) #Lo guardé en un nueva variable para que se 
#actualice y pueda trabajar siempre con el cambio que estoy realizando
```

```python
LibrosArreglo.Ventas_Aproximadas.str.split('[',expand=True) #La columna de ventas aproximadas tiene dos numeros en corchete que 
#son los indices de la tabla en la pagina web, pero aqui no lo toma como tal. Entonces, lo que quiero hacer es separar ##millones
#de los corchetes para luego eliminarlo. Primero separé en dos columnas lo que quiero que se quede y lo que no quiero con la funcion
#split.
```

```python
VentasArreglo=LibrosArreglo.Ventas_Aproximadas.str.split('[',expand=True) #Lo guardé en un nueva variable para que se 
#actualice y pueda trabajar siempre con el cambio que estoy realizando
```

```python
VentasArreglo.iloc[:,0] #Con la funcion .iloc me quedaré con la columna que quiero y eliminaré la que no me sirve
```

```python
VentasFinal=VentasArreglo.iloc[:,0] #Lo guardé en un nueva variable para que se 
#actualice y pueda trabajar siempre con el cambio que estoy realizando
```

```python
LibrosArreglo.assign(Ventas_Aproximadas=VentasFinal) #Ahora la columna que tenia los errores de los indices la reemplazo con la
#que limpie y me quedé en el paso anterior.
```

```python
Libros1=LibrosArreglo.assign(Ventas_Aproximadas=VentasFinal)  #Lo guardé en un nueva variable para que se 
#actualice y pueda trabajar siempre con el cambio que estoy realizando
```

```python
Libros1.Libro.str.split('(',expand=True) #En esta parte, quiero separar el nombre del libro en español con su nombre en otro 
#idioma ya que estaba entre parentesis. Hice lo mismo que con la columna de ventas aproximadas. 
```

```python
LibrosSeparacion=Libros1.Libro.str.split('(',expand=True) #Lo guardé en un nueva variable para que se 
#actualice y pueda trabajar siempre con el cambio que estoy realizando
```

```python
LibrosSeparacion.iloc[:,0] #Con la funcion .iloc me quedaré con cada una de las columnas por separado, este caso con la 0
```

```python
LibrosNombreEspañol=LibrosSeparacion.iloc[:,0] #Lo guardé en un nueva variable para que se 
#actualice y pueda trabajar siempre con el cambio que estoy realizando
```

```python
LibrosSeparacion.iloc[:,1] #Con la funcion .iloc me quedaré con cada una de las columnas por separado, este caso con la 1
```

```python
LibrosNombreIngles=LibrosSeparacion.iloc[:,1] #Lo guardé en un nueva variable para que se 
#actualice y pueda trabajar siempre con el cambio que estoy realizando
```

```python
Libros1.assign(Libro=LibrosNombreEspañol) #Ahora la columna inicial la reemplazo con la que guardé los nombres de los libros 
#en español.
```

```python
Libros2=Libros1.assign(Libro=LibrosNombreEspañol)#Lo guardé en un nueva variable para que se 
#actualice y pueda trabajar siempre con el cambio que estoy realizando
```

```python
Libros2.assign(NombreEnIngles=LibrosNombreIngles) #Ahora agrego una nueva columna que será  la que guardé los nombres de los libros 
#en ingles.
```

```python
Libros3=Libros2.assign(NombreEnIngles=LibrosNombreIngles) #Lo guardé en un nueva variable para que se 
#actualice y pueda trabajar siempre con el cambio que estoy realizando
```

```python
Libros3.NombreEnIngles.str.replace(')', '', regex=True) #En esta parte, en la columna de nombres en ingles, quiero quitar el 
#el parentesis final para que solo queden los nombres.
```

```python
InglesArreglado=Libros3.NombreEnIngles.str.replace(')', '', regex=True)
InglesArreglado #Lo guardé en un nueva variable para que se actualice y pueda trabajar siempre con el cambio que estoy 
#realizando
```

```python
Libros3.assign(NombreEnIngles=InglesArreglado) #Ahora la columna que no estaba limpia la reemplazo con la que guardé los nombres 
#de los libros en ingles u otros idiomas.
```

```python
Libros4=Libros3.assign(NombreEnIngles=InglesArreglado) #Lo guardé en un nueva variable para que se 
#actualice y pueda trabajar siempre con el cambio que estoy realizando
```
